```python
#HW5A.4
a=4
w=0.75
#0-4 v=-wx
#4-8 v=2aw-wx
import matplotlib.pyplot as plt
x_quad = np.linspace(0, 4, 100)
x_linear = np.linspace(4, 8, 100)

def quadratic_func(x):
    return -w*x
def linear_func(x):
    return 2*a*w -w*x
plt.figure(figsize=(8, 6))
plt.plot(x_quad, quadratic_func(x_quad), label='一次函数 ', color='blue')
plt.plot(x_linear, linear_func(x_linear), label='一次函数 ', color='red')
plt.xlabel('x') 
plt.ylabel('f(x)')
plt.title('函数图像') 
plt.grid(True, linestyle='--', alpha=0.7) 
plt.axvline(x=4, color='gray', linestyle=':', alpha=0.5, label='x=L（分界点）')  
plt.legend()
plt.show()



#0-4 m=-wx**2/2
#4-8 m=2aw(x-a)-wx**2/2
x_1 = np.linspace(0, 4, 100)
x_2 = np.linspace(4, 8, 100)
def quadratic_func(x):
    return -w*x**2/2
def linear_func(x):
    return 2*a*w*(x-a)-w*x**2/2
plt.figure(figsize=(8, 6))
plt.plot(x_1, quadratic_func(x_1), label='一次函数 ', color='blue')
plt.plot(x_2, linear_func(x_2), label='一次函数 ', color='red')
plt.xlabel('x') 
plt.ylabel('f(x)')  
plt.title('函数图像')  
plt.grid(True, linestyle='--', alpha=0.7) 
plt.axvline(x=4, color='gray', linestyle=':', alpha=0.5, label='x=L（分界点）')  
plt.legend()
plt.show()
```
