```python
#HW5B.9
import numpy as np

m = 50.00 # kg
mu_A = 0.20 # 
def f(theta):
    return 3*mu_A*np.pi*(1-np.sin(theta))+4*mu_A*np.sin(theta)*np.cos(theta)-4*np.sin(theta)**2
#对水平方向受力分析：f=P*sin(theta)=mu*(G+P*cos(theta))
#力矩：Pr=Psin(theta)*r+G*4r/3/pi*sin(theta)

from scipy.optimize import fsolve
beta_initial = 0.01  # Initial guess in radians
beta_solution = fsolve(f, beta_initial)[0]
beta_degrees = np.degrees(beta_solution)
print("beta =",beta_degrees,"degrees")
p=mu_A*m*9.81/(np.sin(beta_solution)-mu_A*np.cos(beta_solution))
print("p =",p,"N")
```

```python
#HW5B.10
import numpy as np
W = 8000.00 # lb
mu_A = 0.20 # 
mu_B = 0.35 # 
a = 10.00 # ft
b = 15.00 # ft
theta = np.radians(10.00) # deg


fa=W*b/(a+b)
A=np.array([[np.sin(theta),np.cos(theta)-mu_B*np.sin(theta)],
           [np.cos(theta),-np.sin(theta)-mu_B*np.cos(theta)]
           ])
b=np.array([W-fa,mu_A*fa])
x=np.linalg.solve(A,b)
T=x[0]
print("T =",T/1000,"kips")
```
