```python
#HW7A.8
import numpy as np

# 给定参数
R = 8.0  # 外半圆半径 (m)
r = 1.0   # 内半圆半径 (m)
H = 15.0  # 中间矩形高度 (m)

# ---------------------- 外跑道形惯性矩计算 ----------------------
# 1. 中间矩形对截面形心x轴的惯性矩（形心与截面形心重合）
I_outer_rect = (2 * R) * (H ** 3) / 12

# 2. 单个外半圆对截面形心x轴的惯性矩
A_semi_outer = (np.pi * R ** 2) / 2  # 单个半圆的面积
# 半圆对截面形心x轴的惯性矩（展开平行轴定理后的简化形式）
term1_outer = (np.pi * R ** 4) / 8
term2_outer = A_semi_outer * ( (H/2)**2 + (4 * R * H) / (3 * np.pi) )
I_semi_outer = term1_outer + term2_outer

# 3. 外跑道形总惯性矩（矩形 + 2个半圆）
I_outer = I_outer_rect + 2 * I_semi_outer
# ---------------------- 内跑道形惯性矩计算 ----------------------
# 1. 中间小矩形对截面形心x轴的惯性矩
I_inner_rect = (2 * r) * (H ** 3) / 12

# 2. 单个内半圆对截面形心x轴的惯性矩
A_semi_inner = (np.pi * r ** 2) / 2  # 单个小半圆的面积
term1_inner = (np.pi * r ** 4) / 8
term2_inner = A_semi_inner * ( (H/2)**2 + (4 * r * H) / (3 * np.pi) )
I_semi_inner = term1_inner + term2_inner

# 3. 内跑道形总惯性矩（小矩形 + 2个小半圆）
I_inner = I_inner_rect + 2 * I_semi_inner


# ---------------------- 最终惯性矩（外轮廓 - 内轮廓） ----------------------
I_x = I_outer - I_inner

print(f"关于形心x轴的惯性矩 I_x = {I_x:.6f} m^4")
```

```python
#HW7A.9
import numpy as np

t = 30.0 # mm
H = 120.0 # mm
a = 70.0 # mm
L = 150.0 # mm
#需要先确定上翼板的宽度x(不包括t)
def f(x):
    s1=t*(t+x)
    s2=H*t
    s3=t*(t+L)
    x1=(t+x)/2
    x2=t/2
    x3=(t+L)/2
    return (s1*x1+s2*x2+s3*x3)-(s1+s2+s3)*(t+a)

from scipy.optimize import fsolve
x_initial_guess = 100.0
x_solution = fsolve(f, x_initial_guess)
x = x_solution[0]
print(f"上翼板的宽度 x = {x:.2f} mm")


a1=t*(t+x)
y1=t/2+H+t
a2=t*H
y2=t+H/2
a3=t*(t+L)
y3=t/2
y_hat=(a1*y1+a2*y2+a3*y3)/(a1+a2+a3)
print(y_hat)
I1=(t+x)*t**3/12+a1*(y_hat - y1)**2   #I=b*h^3/12 矩形截面绕其水平形心轴（平行于底边的形心主轴） 的惯性矩（截面二次轴矩）公式
I2=t*H**3/12+a2*(y_hat - y2)**2
I3=(t+L)*t**3/12+a3*(y_hat - y3)**2
I=I1+I2+I3
print(I)
```
