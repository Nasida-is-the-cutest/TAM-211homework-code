如果对y方向的积分不好表达，可以使用math 231的知识：y=积分号yy/2*dx
如果看到代码中列了一堆x1,x2，y1,y2之类的东西，说明需要列表格计算

```python
#HW6A.2
import numpy as np
from scipy import integrate
r=93
def f1(x):
    return np.cos(x)
def f2(x):
    return 1
x=r*integrate.quad(f1,-2/3*np.pi,2/3*np.pi)[0]/integrate.quad(f2,-2/3*np.pi,2/3*np.pi)[0]
print(x)
```

```python
#HW6A.6
a=6
h=5
n=3
#x=(a**n/h*(h-y))**1/n
#y=h-(h*x**n/a**n)
def fx(x):
    return (h-(h*x**n/a**n))*x
def f2(x):
    return h-(h*x**n/a**n)
def fy(x):
    return (h-(h*x**n/a**n))**2/2   #来自math的知识：y=积分 yy/2*dx
x=integrate.quad(fx,0,a)[0]/integrate.quad(f2,0,a)[0]
y=integrate.quad(fy,0,a)[0]/integrate.quad(f2,0,a)[0]
print(x,y)
```
