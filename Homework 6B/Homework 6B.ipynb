```python
import numpy as np
from scipy.optimize import fsolve

# 已知参数
r2 = 1.9  # rho2/rho1
r1 = 1
g = 9.81
theta = np.radians(17)
h1 = 5    # B点在流体1中的深度
h2 = 4    # AB杆的竖直高度
L = h2 / np.sin(theta)  # AB杆长度（沿杆方向）


# 步骤1：计算流体1对B点的力矩M1
# 流体1的压力（垂直于AB杆）
F1 = r1 * g * (h1 * L + np.sin(theta) * L**2 / 2)
# 流体1的压力中心（沿AB杆到B的距离）
numerator_s1 = h1 * L**2 / 2 + np.sin(theta) * L**3 / 3
denominator_s1 = h1 * L + np.sin(theta) * L**2 / 2
s1 = numerator_s1 / denominator_s1
M1 = F1 * s1  # 流体1对B的力矩


# 步骤2：定义流体2的力矩函数（平衡时M2=M1）
def fun(H):
    p2_B = H - h2  # 流体2中B点的深度（H是流体2总高度）
    # 流体2的压力（垂直于AB杆）
    F2 = r2 * g * (p2_B * L + np.sin(theta) * L**2 / 2)
    # 流体2的压力中心（沿AB杆到B的距离）
    numerator_s2 = p2_B * L**2 / 2 + np.sin(theta) * L**3 / 3
    denominator_s2 = p2_B * L + np.sin(theta) * L**2 / 2
    s2 = numerator_s2 / denominator_s2
    M2 = F2 * s2  # 流体2对B的力矩
    return M2 - M1  # 平衡条件：M2 = M1


# 步骤3：求解最大高度Hmax
H_max = fsolve(fun, x0=4)  # 初始猜测H=4
print("H_max =", H_max[0])  # 输出：6.44444444444
```
